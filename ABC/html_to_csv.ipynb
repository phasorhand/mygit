```python
# encoding=utf-8
from scrapy.selector import Selector
import re
import xlwt

with open('C:\\Users\\Administrator\\Desktop\\mine\\lj\\word_to_excel\\1.html', 'rb') as fobj:
    html_doc = fobj.read()

    single_info = {}
    # 姓名
    name = Selector(text=html_doc).xpath(u'//span/text()').extract()[0]
    print 'name',name
    single_info['name'] = name

    # 年龄及性别
    age_sex = Selector(text=html_doc).xpath(u'//span/text()').extract()[1]
    # print 'age_sex',age_sex
    age = re.search(r'[1-9]{2}',age_sex)
    if age is not None:
        print 'age', age.group()
        single_info['age'] = age.group()
    sex = re.search(u'男|女', age_sex)
    if sex is not None:
        print 'sex', sex.group()
        single_info['sex'] = sex.group()

    # 志愿
    job_expectancy_string = Selector(text=html_doc).xpath(u'//tr[child::td[contains(text(),"期望职业")]]/td[last()]//text()').extract()[1]
    # print 'job_expectancy_string', job_expectancy_string
    job_expectancys = re.findall(u'[\u4e00-\u9fa5/]+',job_expectancy_string)
    i = 0;
    job_expectancy_dict = {}
    for job_expectancy in job_expectancys:
        if job_expectancy is not None:
            if i == 0:
                job_expectancy_name = 'first_job_expectancy'
            elif i == 1:
                job_expectancy_name = 'second_job_expectancy'
            else:
                job_expectancy_name = 'other_job_expectancy'
            single_info[job_expectancy_name] = job_expectancy
            print job_expectancy_name,job_expectancy
        i += 1

    # 手机号
    phone_number = Selector(text=html_doc).xpath(u'//td[contains(text(),"手机")]/text()').extract()[0][-11:]
    if phone_number:
        print 'phone_number', phone_number
        single_info['phone_number'] = phone_number

    # 邮箱
    email_str = Selector(text=html_doc).xpath(u'//td[contains(text(),"邮箱")]/text()').extract()[0]
    if email_str:
    #     print 'email_str', email_str
        email_re_result = re.search('[\w\d]+@[\w\d]+\.com', email_str)
        if email_re_result:
            email = email_re_result.group()
            single_info['email'] = email
        print 'email:', email

    # 本科
    bechalor_info = Selector(text=html_doc).xpath(u'//td[contains(text(),"本科")]/text()/preceding::*').extract()[-1]
    if bechalor_info:
    #     print 'bechalor_info:', bechalor_info
        bechalor_info_list = re.findall(u'[\u4e00-\u9fa5]+', bechalor_info)
        if bechalor_info_list:
    #         print 'bechalor_info_list:', bechalor_info_list
            for bechalor_some_info in bechalor_info_list:
    #             print 'bechalor_info:', bechalor_some_info
                if re.search(u'大学|学院',bechalor_some_info):
                    bechalor_school = bechalor_some_info
                    single_info['bechalor_school'] = bechalor_school
                    print 'bechalor_school:', bechalor_school
                else:
                    bechalor_major = bechalor_some_info
                    single_info['bechalor_major'] = bechalor_major
                    print 'bechalor_major:', bechalor_major
    # 硕士 master
    master_info = Selector(text=html_doc).xpath(u'//td[contains(text(),"硕士")]/text()/preceding::*').extract()[-1]
    if master_info:
    #     print 'bechalor_info:', bechalor_info
        master_info_list = re.findall(u'[\u4e00-\u9fa5]+', master_info)
        if master_info_list:
    #         print 'bechalor_info_list:', bechalor_info_list
            for master_some_info in master_info_list:
    #             print 'bechalor_info:', bechalor_some_info
                if re.search(u'大学|学院',master_some_info):
                    master_school = master_some_info
                    single_info['master_school'] = master_school
                    print 'master_school:', master_school
                else:
                    master_major = master_some_info
                    single_info['master_major'] = master_major
                    print 'master_major:', master_major
    # try:
    # info_excel= xlrd.open_workbook('C:\\Users\\Administrator\\Desktop\\mine\\lj\\word_to_excel\\info.xlsx')
    info_excel = xlwt.Workbook(encoding='utf-8')
    book = info_excel.add_sheet('info', cell_overwrite_ok=True)
    # 总的简历数
    for i in range(1):
        print 'i:', i
        try:
            if single_info['first_job_expectancy']:
                print 'first_job_expectancy'
                book.write(i, 3, single_info['first_job_expectancy'])
            if single_info['second_job_expectancy']:
                print 'second_job_expectancy'
                book.write(i, 5, single_info['second_job_expectancy'])
            if single_info['name']:
                print 'name'
                book.write(i, 6, single_info['name'])
            if single_info['sex']:
                print 'sex'
                book.write(i, 7, single_info['sex'])
            if single_info['phone_number']:
                print 'phone_number'
                book.write(i, 8, single_info['phone_number'])
            if single_info['bechalor_major']:
                print 'bechalor_major'
                book.write(i, 9, u'本科')
                book.write(i, 10, single_info['bechalor_major'])
            if single_info['bechalor_school']:
                print 'bechalor_school'
                book.write(i, 11, single_info['bechalor_school'])
            # 博士未考虑
            if single_info['master_major']:
                print 'master_major'
                book.write(i, 9, u'硕士')
                book.write(i, 13, single_info['master_major'])
            if single_info['master_school']:
                print 'master_school'
                book.write(i, 14, single_info['master_school'])
        except Exception as e:
            print 'e:',e
    info_excel.save('info1.xls')
    # except Exception as e:
    #     print 'exception', e
```

```python
import xlrd
import xlwt
from datetime
 
def testXlrd(filename):
    book=xlrd.open_workbook(filename)
    sh=book.sheet_by_index(0)
    print "Worksheet name(s): ",book.sheet_names()[0]
    print 'book.nsheets',book.nsheets
    print 'sh.name:',sh.name,'sh.nrows:',sh.nrows,'sh.ncols:',sh.ncols
    print 'A1:',sh.cell_value(rowx=0,colx=1)
    #如果A3的内容为中文
    print 'A2:',sh.cell_value(0,2).encode('gb2312')
 
def testXlwt(filename):
    book=xlwt.Workbook(encoding='utf-8')
    sheet1=book.add_sheet('hello')
    book.add_sheet('word')
    sheet1.write(0,0,'hello')
    sheet1.write(0,1,'world')
    row1 = sheet1.row(1)
    row1.write(0,'A2')
    row1.write(1,'B2')
     
    sheet1.col(0).width = 10000
     
    sheet2 = book.get_sheet(1)
    sheet2.row(0).write(0,'Sheet 2 A1')
    sheet2.row(0).write(1,'Sheet 2 B1')
    sheet2.flush_row_data()
     
    sheet2.write(1,0,'Sheet 2 A3')
    sheet2.col(0).width = 5000
    sheet2.col(0).hidden = True
     
    book.save(filename)
 
if __name__=='__main__':
    testXlrd(u'你好。xls')
    testXlwt('helloWord.xls')
    base=datetime.date(1899,12,31).toordinal()
    tmp=datetime.date(2013,07,16).toordinal()
    print datetime.date.fromordinal(tmp+base-1).weekday()
```
